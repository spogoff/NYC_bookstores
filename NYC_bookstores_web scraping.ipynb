```python
#import dependencies
import requests
import urllib.request
import time
from bs4 import BeautifulSoup
import re
import pandas as pd
```

```python
#request url
url = 'https://www1.nyc.gov/site/mome/industries/nyc-bookstores.page'
response = requests.get(url)
```

```python
# set and parse the html object 
soup = BeautifulSoup(response.text, 'html.parser')
```

```python
# get all results from thml
links_soup = soup.select('ul li a')
name_add_soup = soup.select('ul li')
```

```python
# make a list of names and addresses
all_names_adds = []
for tag in name_add_soup:
    all_names_adds.append(tag.get_text())
all_names_adds
```

```python
# make a list of all links
all_links = []
for tag in links_soup:
    all_links.append(tag.get('href'))
all_links
```

```python
# Clean links list
all_links = [x for x in all_links[13:]]
all_links
```

```python
# Clean names_adds list 
all_names_adds = [x for x in all_names_adds[14:]]
all_names_adds
```

```python
# seperate names and addresses
all_names = []
all_addresses = []
for item in all_names_adds:
    name, address = item.split(' — ',1)
    all_names.append(name)
    all_addresses.append(address)
```

```python
nyc_bookstores_df = pd.DataFrame({'Name':all_names,'Address':all_addresses,'Link':all_links})
nyc_bookstores_df
```

```python
# Save dataframe as csv
nyc_bookstores_df.to_csv('nyc_bookstores.csv',index=False)
```
