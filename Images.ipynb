```python
#import dependencies
import requests
import urllib.request
import time
from bs4 import BeautifulSoup
import re
import pandas as pd
from splinter import Browser 
```

```python
# read csv file as dataframe
nyc_bookstores_df = pd.read_csv('nyc_bookstores_location.csv')
nyc_bookstores_df
```

```python
# Create a new column for img urls
nyc_bookstores_df.loc[:,'Image URL'] = 'Not Found'
nyc_bookstores_df
```

```python
# Scrape img urls
img_urls = []
for row in range(0,68):
    try:
        url = nyc_bookstores_df['Link'][row]
        response = requests.get(url)
        soup = BeautifulSoup(response.text, 'html.parser')
        img_url = soup.find('img')['src']
        nyc_bookstores_df['Image URL'][row] = img_url
#         img_urls.append(img_url)
        time.sleep(1.5)
    except:
        print(url)
```

```python
# Temp 
## Manually fix the missing img_urls
nyc_bookstores_df.loc[nyc_bookstores_df['Link'] == 'https://www.albertine.com/holiday-2020/', "Image URL"] = "https://www.albertine.com/wp-content/themes/albertine/assets/images/albertine.png"
nyc_bookstores_df.loc[nyc_bookstores_df['Link'] == 'https://aperture.org/shop', "Image URL"] = "https://aperture.org/wp-content/uploads//2020/06/2012_aperture_logotype_cover.svg"
nyc_bookstores_df.loc[nyc_bookstores_df['Link'] == 'https://lalibrairiedesenfants.nyc/', "Image URL"] = "https://lalibrairiedesenfants.nyc/wp-content/uploads/2018/12/HOME-BANNER-plain-200x123.png"
nyc_bookstores_df.loc[nyc_bookstores_df['Link'] == 'http://www.strandbooks.com/', "Image URL"] = "https://d2p7wwv96gt4xt.cloudfront.net/S/static/strand-logo-exact.png"
nyc_bookstores_df.loc[nyc_bookstores_df['Link'] == 'https://www.toposbookstore.com/', "Image URL"] = "https://www.toposbookstore.com/uploads/b/b1a882f0-cfdc-11e9-b60c-eb707208c907/toposFB_sm.gif?width=400"
```

```python
# # # Clean img_urls    
full_img_urls = []
for index in range(0,68):
    full_img_url = ""
    img_url = nyc_bookstores_df['Image URL'][index]
    link = nyc_bookstores_df['Link'][index]
    if link.endswith("/"):
        link = link
    else:
        link = link+"/"
# If image url starts with http then keep it as is
    if img_url.startswith("http"):
        full_img_url = f'{img_url}'
#         print('good to go')
# If not, then check for extra slashes
    else:
        if img_url.startswith("//"):
            full_img_url = f'{link}{img_url[2:]}'  
        elif img_url.startswith("/"):
            full_img_url = f'{link}{img_url[1:]}'
#         print('needed fixing')
    full_img_urls.append(full_img_url)

nyc_bookstores_df['Full Image URL'] = full_img_urls
```

```python
# Save dataframe as csv
nyc_bookstores_df.to_csv('nyc_bookstores_location_images.csv',index=False)
```
