```python
# Import dependencies
import gmaps
import pandas as pd
from PIL import Image
# Import the API key.
from config import g_key
```

```python
# read csv file as dataframe
nyc_bookstores_df = pd.read_csv('nyc_bookstores_location_images.csv')
nyc_bookstores_df
```

```python
# Configure gmaps to use your Google API key.
gmaps.configure(api_key=g_key)
```

```python
info_box_template = """
<dl>
<dt>{Name}</dt><dd></dd>
<a href={Link}>Visit Website</a>
<img src={Full Image URL}/>
</dl>
"""
```

```python
# Store the DataFrame Row.
popup_info = [info_box_template.format(**row) for index, row in nyc_bookstores_df.iterrows()]
popup_info
```

```python
# Map
# Get the latitude and longitude.
locations = nyc_bookstores_df[["Latitudes","Longitudes"]]
new_york_coordinates = (40.75, -74.00)
fig = gmaps.figure(center=new_york_coordinates, zoom_level=10.5)
marker_layer = gmaps.marker_layer(locations, info_box_content = popup_info)
# bookstore_layer = gmaps.symbol_layer(locations, fill_color='blue', stroke_color='green', scale=3, info_box_content = popup_info)
fig.add_layer(marker_layer)
# Call the figure to plot the data.
fig
```
